```python
#Get the dataset
import pandas as pd
df = pd.read_csv("new_rainfall.csv")
df.head(10)
```

```python
#Check number of missing values
df.isnull().sum()
```

```python
#Generate missing values matrix
import numpy as np
nan_mat = np.random.random(df.iloc[:,2:].shape)<0.04
nan_mat.sum()
```

```python
#Add missing values
df1 = df.iloc[:,2:].mask(nan_mat)
dirty = pd.concat([df.iloc[:,:2],df1],axis = 1)
dirty.head(10)
```

```python
#Check number of missing values
dirty.isnull().sum()
```

```python
# 4.1% missing values inserted
```

```python
#Load missing values to a new file
dirty.to_csv("dirty_rainfall.csv")
```
